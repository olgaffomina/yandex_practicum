### Технологический процесс

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

Опишем каждую стадию: 
 - 1. Флотация
 
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
 - 2. Очистка 
 
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.


### Описание данных

**Технологический процесс**
 - Rougher feed — исходное сырье
 - Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
 - Xanthate **— ксантогенат (промотер, или активатор флотации);
 - Sulphate — сульфат (на данном производстве сульфид натрия);
 - Depressant — депрессант (силикат натрия).
 - Rougher process (англ. «грубый процесс») — флотация
 - Rougher tails — отвальные хвосты
 - Float banks — флотационная установка
 - Cleaner process — очистка
 - Rougher Au — черновой концентрат золота
 - Final Au — финальный концентрат золота

**Параметры этапов**
 - air amount — объём воздуха
 - fluid levels — уровень жидкости
 - feed size — размер гранул сырья
 - feed rate — скорость подачи


**Наименование признаков**

Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]
Пример: rougher.input.feed_ag

**Возможные значения для блока [этап]:**
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

**Возможные значения для блока [тип_параметра]:**
 - input — параметры сырья
 - output — параметры продукта
 - state — параметры, характеризующие текущее состояние этапа
 - calculation — расчётные характеристики

### Проверка эффективности обогащения

 Для начала проверим, что эффективность обогащения рассчитана правильно. Вычислим ее на обучающей выборке для признака `rougher.output.recovery`. И найдем MAE между получившимися расчётами и значением признака.


Эффективность обогащения рассчитывается по формуле:

$$recovery=\frac{C(F-T)}{F(C-T)})*100\%$$

где:
 - C — доля золота в концентрате после флотации/очистки;
 - F — доля золота в сырье/концентрате до флотации/очистки;
 - T — доля золота в отвальных хвостах после флотации/очистки.


```python
#проверка на наличие пропусков в необходимых строках(для корректности расчета эффект. обогащения)
train_recovery = train[['rougher.output.concentrate_au','rougher.input.feed_au','rougher.output.tail_au','rougher.output.recovery']]
print(train_recovery.isnull().sum())
```

```python
#введем переменные
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']

#Расчет recovery по формуле
recovery = C*(F-T)/(F*(C-T))*100
print(recovery)
```

```python
#расчет среднего абсолютного отклонения расчитанных данных от значения признака в отчете 
mae_recovery = mean_absolute_error (train['rougher.output.recovery'],recovery)
print('MAE = ',mae_recovery)
```

```python
#просмотр данных в rougher.output.recovery
print(train['rougher.output.recovery'].describe())
print()
print('Количество строк со значением 0 в столбце rougher.output.recovery: ', len(train[train['rougher.output.recovery'] == 0]))
```

**Вывод:**

Среднее абсолютное отклонение расчетных показателей от значения признака `rougher.output.recovery` составляет  9.735 (МАЕ = 9.735). 

Погрешность минимальная, значение коэффициента rougher.output.recovery в датафрейме посчитаны верно.

### Функция для вычисления итоговой sMAPE

Формула для расчета sMAPE:
$$ sMAPE = \frac{1}{N}\sum_{i=1}^N\frac{|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i})/2}*100\% $$ 

где 

\\( y_i \\) - значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество 

\\( \hat{y_i} \\) - значение предсказания для объекта с порядковым номером i, например, в тестовой выборке


\\( N \\) - количество объектов в выборке

\\( \sum_{i=1}^N \\) - суммирование по всем объектам выборки (i меняется от 1 до N)

```python
#функция для вычисления smape
def smape(targets, predictions):
    fraction = abs(targets - predictions) / ((abs(targets) + abs(predictions)) / 2)
    sum_fraction = fraction.sum()
    smape = (1 / len(targets)) * sum_fraction * 100
    return smape
```

Итоговая метрика складывается из двух величин:
$$sMAPE(итоговое) = 25\% * sMAPE(rougher)+75\% * sMAPE(final)$$

```python
def final_smape_calc(smape_rougher, smape_final):
    final_smape = 0.25 * smape_rougher + 0.75 * smape_final
    return final_smape
```

Создадим параметры для будущих моделей 

```python
#параметры для проведения GridSearchCV (подбор лучших параметров)
parameters_LR = {}

parameters_DT = {'decisiontreeregressor__max_depth': range(1, 11)}

parameters_RF = {'randomforestregressor__max_depth': list(range(2, 12, 2)),
                 'randomforestregressor__n_estimators': list(range(10,60,10))
                }

model_LR = make_pipeline(StandardScaler(), LinearRegression())
model_DT = make_pipeline(StandardScaler(), DecisionTreeRegressor(random_state=12345))
model_RF = make_pipeline(StandardScaler(), RandomForestRegressor(random_state=12345))
```

 **!Спойлер:**
при добавлении третьего гиперпараметра (min_samples_leaf) в случайный лес - время обучения увеличивается с 14 минут до 1часа 32 мин, при этом значение final sMAPE изменилось в 4 цифре после запятой(при моем округлении в 2 цифры после запятой разница не выявлется). Приняв волевое решение - удалила третий гиперпараметр и оставила только два стандартных


```python
#функция для расчета final sMAPE
#на вход: модель, features, target, параметры модели
#на выход: итоговое значение sMape и параметры лучшей модели
def grid_search_cv(model, features_rougher, features_final, target_rougher, target_final, parameters):

    limit_smape = 1000
    
    scorer = make_scorer(smape, greater_is_better=False)
    
    #подбор параметров модели для этапа флотации 
    grid_rougher = GridSearchCV(estimator=model, param_grid=parameters, cv=5, scoring=scorer)
    #запуск обучения
    grid_rougher.fit(features_rougher, target_rougher)
    #лучшее значение smape этапа флотации(rougher)
    smape_rougher = grid_rougher.best_score_
     
    #подбор параметров, моделирование и расчет sMAPE для финального этапа(final)   
    grid_final = GridSearchCV(estimator=model, param_grid=parameters, cv=5, scoring=scorer)
    grid_final.fit(features_final, target_final)
    smape_final = grid_final.best_score_

    final_smape = final_smape_calc(smape_rougher, smape_final)
    
    if final_smape < limit_smape:
        best_model_rougher = grid_rougher.best_estimator_
        best_model_final = grid_final.best_estimator_
        best_smape_final = final_smape
# параметр greater_is_better=False определяет функцию минимизации решения (low is good), 
#и возвращает решение с противоположным знаком, поэтому умножаем на -1.
        print(f'Итоговое sMAPE:', round(best_smape_final*(-1), 2))  
        print("Параметры лучшей модели для rougher:", grid_rougher.best_params_)
        print("Параметры лучшей модели для final:", grid_final.best_params_)
        print()
    return best_model_rougher, best_model_final
```
