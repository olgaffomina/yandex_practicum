### Удаление пропусков

```python
data.isna().sum()
```

 В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Категоризация данных

На основании диапазонов, указанных ниже, создадим в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

```python
#уникальные цели кредита
data['purpose'].unique()
```

 Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

## Сущетвует ли зависимость между количеством детей и возвратом кредита в срок?

```python
display(data.head())
```

```python
#  функция для формирования сводной таблицы, с переименованием столбцов и добавления расчета в процентах

def pvt(index, columns):
    data_pvt = data.pivot_table(index = index,
                            columns = columns,
                            values = 'total_income',
                            aggfunc = 'count',fill_value=0,
                            margins = True)
    
    data_pvt = data_pvt.rename(columns={0:"without_dept",1:"in_dept"})

    data_pvt['without_dept_%'] = (data_pvt ['without_dept']/data_pvt['All']*100).round(2)
    data_pvt['in_dept_%'] = (data_pvt ['in_dept']/data_pvt['All']*100).round(2)
    return data_pvt

    
data_pivot_status = pvt('children','debt')
display(data_pivot_status)
```

**Вывод:**
 Наиболее ответсвенные,согласно отчету, бездетные клиенты. Процент выплаты кредита в срок составляет  92.46%, в противовес им можно выделить клиентов с двумя детьми (90,55%) и с одним ребенком (90,77%).Тем самым можно опровергнуть стереотип, что бездетные клиенты менее ответственны.
