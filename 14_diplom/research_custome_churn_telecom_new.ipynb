## Описание проекта: 


Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. 

**Цель проекта:** спрогнозировать уход клиента и определить факторы, влияющие на отток клиентов

**Описание услуг**

Оператор предоставляет два основных типа услуг: 

 - Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
 - Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
 
Также доступны такие услуги:

 - Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
 - Выделенная линия технической поддержки (TechSupport);
 - Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
 - Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).
 
 
За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Описание данных**

Данные состоят из файлов, полученных из разных источников:

 - **contract_new** — информация о договоре;
 - **personal_new** — персональные данные клиента;
 - **internet_new**— информация об интернет-услугах;
 - **phone_new** — информация об услугах телефонии.

Информация о договорах актуальна на 1 февраля 2020.


**Данные файлов:**


**contract_new**

- `customerID` - ID клиента
- `BeginDate` - дата заключения договора
- `EndDate` - дата расторжения договора
- `Type` - тип оплаты(ежемесячная оплата,договор годовой или двухгодовой)
- `PaperlessBilling` - электронный счет(да/нет)
- `PaymentMethod` - метод оплаты (Electronic check/Mailed check/Bank transfer (automatic)/Credit card (automatic))
- `MonthlyCharges` - ежемесячная плата
- `TotalCharges` - полная плата



 **personal**
 
 - `customerID` - ID клиента
 - `gender` - пол
 - `SeniorCitizen` - является ли клиент пенсионером
 - `Partner` - наличие супруга/супруги
 - `Dependents` - наличие иждивенцов
 

**internet_new**
- `customerID` - ID клиента
- `InternetService` - тип подключения интернета (DSL, оптоволокно)
- `OnlineSecurity` -   наличие услуги облачного хранилище файлов для резервного копирования данных
- `DeviceProtection` - услуга антивирус 
- `TechSupport` - наличие выделенной линии технической поддержки 
- `StreamingTV` - услуга стримингового телевидения
- `StreamingMovies` -  наличие услуги каталога фильмов 


**phone_new**

- `customerID` - ID клиента
- `MultipleLines` - наличие нескольких телефонных линий у клиента

### Подготовка данных

План действий:

 - проверка на аномалии
 - проверка типов данных
 - генерация новых столцов( количество доп услуг, срок действия договора в днях)

```python
#замена пробелов на 0 в total_charges
full_data['total_charges']= full_data['total_charges'].replace(' ',0)
#проверка
full_data[full_data['total_charges'] ==' ']
```

Заменим тип данных 'total_charges' на числовой

```python
#замена типа данных 'total_charges' на int
full_data['total_charges'] = pd.to_numeric(full_data['total_charges'])
```

Преобразуем  к бинарной класификации с типом object по аналогии с другими полями

```python
#преобразование senior_citizen в тип данных object
full_data['senior_citizen'] = full_data['senior_citizen'].replace(1, 'Yes').replace(0, 'No').astype('object')
```

Следующим шагом избавимся от пустых значений, рассмотрим подробнее эти данные

```python
#просмотр инфо строк с пустыми значениями internet_service 
display(full_data.query('internet_service.isna()').head(10))
```

Сответственно, отсутвующие данные в столбце multiple_lines говорит, что клиент использует только интернет. Заменим эти пропуски на 'only_internet'

 Где отсутствуют данные в столцах internet_service клиенты пользуются только телефоннй связью. Заменим пустые сроки на 'only_phone'
 
 Все пустые строки в доп услугах online_security, online_backup, device_protection, tech_support, streaming_tv, streaming_movies - заменим на 'no_internet'

```python
#замена пустых строк в multiple_lines на 'only_internet' (клиенты пользуются только интернетом)
full_data['multiple_lines'] = full_data['multiple_lines'].fillna('only_internet')

#замена пустых строк в internet_service на 'only_phone' (клиенты пользуются только стационарным телефоном)
full_data['internet_service'] = full_data['internet_service'].fillna('only_phone')

# замена пустых строк в online_security, online_backup, device_protection, tech_support, streaming_tv ,streaming_movies
# на ' no_internet'(клиенты пользуются только стационарным телефоном)
nan_list = ['online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 
            'streaming_movies']

for col in nan_list:
    full_data[col] = full_data[col].fillna('no_internet')
    
    
#проверка
full_data.info()
```

Все строк заполнены, добавление данных произошло успешно. Сгенерируем дополнительно столбец с информацией, какие услуги использует клиент(internet/phone/both)

```python
# добавление столбца с типом услуг(internet/phone/both)
def service_gr(row):
    if  row['multiple_lines'] == 'only_internet':
        return 'only_internet'
    elif row['internet_service'] == 'only_phone':
        return 'only_phone'
    else:
        return 'both'

full_data['service_type'] = full_data.apply(service_gr, axis=1)

#проверка
print(full_data[['multiple_lines','internet_service','service_type']].head(10))
```

Сгенерируем дополнительные данные 'total_days' со сроком действия договора(в днях). Так как данные на  2020-02-01, рассчитаем срок действия договара на это число. Предварительно заменим данные в 'begin_date' и 'end_date' в datetime ( для возможности провести расчет) и заменим пустые строки в 'end_date' на 2020-02-01(это говорит, что договор еще действителен)

```python
#замена пустых
full_data['end_date'] = full_data['end_date'].replace('No','2020-02-01')
#проверка
print(full_data['end_date'].unique())
```

```python
#замена типа данных в 'begin_date' и 'end_date' на  datetime
full_data['begin_date'] = pd.to_datetime(full_data['begin_date'])
full_data['end_date'] = pd.to_datetime(full_data['end_date'])


#подсчет срока действия договора (в днях)
full_data['lifetime_days'] = (full_data['end_date']-full_data['begin_date']).dt.days

#подсчет срока действия договора (в месяцах)
full_data['lifetime_month'] = ((full_data['end_date']-full_data['begin_date'])/np.timedelta64(1,'M')).astype(int)

#проверка
display(full_data[['begin_date','end_date','lifetime_month' ,'lifetime_days']].head(10))
```

Посмотрим как признаки влияют на целевой признак 'exited'. Построим круговые диаграммы для наглядного отображения.
Вызовем функцию для всех категориальных столбцов. На вход - столбец, на выходе сводная таблица и круговая диаграмма

```python
#загрузка библиотек для проверки гипотез 
from scipy.stats import chi2, chi2_contingency
```

```python
#категориальные признаки
cols_cat = ['type', 'paperless_billing', 'payment_method', 'gender', 'senior_citizen', 'dependents', 'partner',
           'dependents', 'internet_service','multiple_lines', 'service_type']

for col in cols_cat:
    #сводная таблица
    pivot = full_data.replace({1 : 'клиент ушел', 0 : 'действующий клиент'}).pivot_table(index = col,
                                                                                     columns = 'exited',
                                                                                     values = 'customer_id',
                                                                                     aggfunc = 'count')
    
    print('\033[1m' + 'распределение клиентов по признаку: '+ col +'\033[0m')
    display(pivot)# вывод таблицы
    
    #проверка гипотез вляния признака на целевую
    chi2_stat, p, dof, expected = chi2_contingency(pivot.values)
    print('p-value=', p)

   # интерпретируем значение статистики
    critical = chi2.ppf(0.95, dof)

    if abs(chi2_stat) >= critical:
        print('Гипотеза об отсутствии различий между группами отвергается. Признак значимый.')
    else:
        print('Гипотеза об отсутствии различий между группами подтверждается. Признак не значимый')
        

    # интерпретируем значение p-value
    alpha = 0.05
    if p <= alpha:
        print('Гипотеза об отсутствии различий между группами отвергается. Признак значимый.')
    else:
        print('Гипотеза об отсутствии различий между группами подтверждается. Признак не значимый')

    #построение графика
    plt.figure(figsize=(8, 6))
    
    plt.subplot(1, 2, 1)
    plt.pie(pivot['клиент ушел'], autopct='%1.2f%%')
    plt.legend(labels=pivot.index, loc='best',borderaxespad=0)
    
    plt.title('Ушедшие клиенты')
   
    plt.subplot(1, 2, 2)
    plt.pie(pivot['действующий клиент'], autopct='%1.2f%%')
    plt.title('Оставшиеся клиенты')
    plt.show()
```

**Краткий вывод:** 
Наибольшее количество "сохранившихся" клиентов с ежемесячной оплатой, в одинаковой пропорции ушли клиенты  с ежемесячной,годовой и 2хгодовой оплатой.

Все гипотезы о значимости признака и влияния на целевую подтвердились, помимо одного признака - пола клиента(gender)

Для кодирования и масштабирования, используем **make_column_transformer**:

```python
#кодирование (OneHotEncoder) и маштабирование (StandardScaler())
#drop='first' удаляет первый признак из закодированных:
#таким образом обходим dummy-ловушку
#задаём handle_unknown='ignore':
#игнорируется ранее невстречающиеся значения признака (при transform)

col_transformer_lg = make_column_transformer(
    (OneHotEncoder(drop='first',handle_unknown='ignore',sparse=False),ohe_features ),
    (StandardScaler(), num_features),
    verbose_feature_names_out=False)
```

```python
#создание копии features_train
features_train_boost = features_train.copy()
#список категориальных столбцов
categorical = features_train.select_dtypes(include='object').columns.to_list()
#преобразование в тип данных 'category'
for col in categorical:
    features_train_boost[col] = features_train_boost[col].astype('category')
#проверка
features_train_boost.info()
```

```python
%%time
#random_state не перебирается, задаём его прямо в модели
model_lg = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')
#словарь с гиперпараметрами и значениями, которые хотим перебрать

param_grid_lg = {"C":[0.1, 1.0],
                'solver': ['lbfgs', 'sag'],
                 'penalty': ['l1', 'l2']}

#подбор гиперпараметры с помощью GridSearchCV
gs_lg = GridSearchCV(estimator=model_lg, param_grid=param_grid_lg, cv=5, scoring='roc_auc', n_jobs=-1)

gs_lg.fit(X_train, target_train)

ROC_AUC_lg = gs_lg.best_score_
best_params_lg = gs_lg.best_params_
best_model_lg = gs_lg.best_estimator_

# лучшее значение ROC-AUC на кросс-валидации
print(f'ROC-AUC best_score: {ROC_AUC_lg}')
# лучшие гиперпараметры
print(f'best_params: {best_params_lg}')
print(f'best_model: {gs_lg.best_estimator_}')

```

### Построение ROC кривой

```python

#построение ROC кривой и расчет AOC-ROC
def roc(model): 
    probabilities_test = model.predict_proba(features_test_boost) #значение вероятностей
    probabilities_one_test = probabilities_test[:, 1] #вероятности со значением 1
    
    fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)
    plt.figure(figsize=(4,4))
    plt.plot(fpr, tpr) #построение графика 
    plt.plot([0,1],[0,1],linestyle = '--') #ROC-кривая случайной модели
    plt.xlim([0.0, 1.0]) #установка грани осей от 0 до 1
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show() 
    auc_roc = roc_auc_score(target_test, probabilities_one_test)
    print("AUC_ROC:", auc_roc)
    return auc_roc
```

```python
#ROC кривая для модели LightGBM
roc(best_model_lgbm)
```

**Пояснения**
ROC-кривая – это кривая вероятности, которая отображает отношение TPR(верно спрогнозированный ушедших клиентов) к FPR (прогноз, что клиент уйдет, а он остался) при различных пороговых значениях и по существу отделяет «сигнал» от «шума». Площадь под кривой (AUC) является мерой способности классификатора различать классы. Чем выше площадь под кривой, тем лучше производительность модели.

Этот график показывает дилемму выбора между верно квалифицированными ушедшими клиентами и верно квалифицированными оставшимися клиентами.
Согласно графику, если устанавливаем низкий порог отсечения (объекты с оценками выше порога считаем принадлежащими классу 1, остальные – 0) модель с высокой точностью будет  прогнозировать ушедших клиентов, но при этом и количество ложно ушедших клиентов увеличится (которые не собирались уходить) и тем самым риск  увеличить расходы на сохранения клиентов.
И наоборот, при высоком пороге низкая доля верно классифицированных ушедших клиентов, но высокая доля верно классифицированных оставшихся клиентов, то есть есть риск потерять клиентов, но при этом сократить расходы на сохранение клиентов
